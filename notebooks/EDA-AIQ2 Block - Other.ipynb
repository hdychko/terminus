```python
import os
import pandas as pd
import numpy as np
```

```python
DATA_PATH = os.path.join(os.getcwd(), '..')
DATA_PATH = os.path.join(DATA_PATH, 'data')
COMMERCIAL_DATA_PATH = os.path.join(DATA_PATH, 'commercial')
```

```python
# read data
df_final = pd.read_csv(os.path.join(DATA_PATH, 'Final_Dataset.csv'), low_memory=False)
df_aiq2 = pd.read_csv(os.path.join(DATA_PATH, 'AIQ2 Block - Cashplus.csv'))
```

```python
print('Final; Number of rows: %.0f; columns: %.0f' % df_final.shape)
print('AIQ 2; Number of rows: %.0f; columns: %.0f' % df_aiq2.shape)
```

```python
# merge datasets
df = pd.merge(df_final, df_aiq2, left_on='UNIQUE_ID', right_on='UNIQUEID', how='outer')
```

```python
print('Merged; Number of rows: %.0f; columns: %.0f' % df.shape)
```
